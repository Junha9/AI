```python
# import statements
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.inspection import permutation_importance
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
# download and split the data
wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol', 'sugar', 'pH']]
target = wine['class']
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
# Let's create RandomForestClassifier
rf = RandomForestClassifier(n_jobs = -1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# Let's figure out the feature importances of the forest
rf.fit(train_input, train_target)
print(rf.feature_importances_)
```

```python
# we can use the out of bag sample by setting oob_score=True. This way out of bag sample is considered as validation set.
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

```python
# Let's use another ensemble alogrithm, its ExtraTressCalssifier which random splitter is applied.
et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# Let's see the feature importances for the extra tree as well.
et.fit(train_input, train_target)
print(et.feature_importances_)
```

```python
# There is also also an ensemble algorithm which uses graident descent. It is called GradientBoosting.
gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# It is super strong for avoiding overfitting. Therefore we can increase the number of tree and learning rate a bit.
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# Seems not too much overfitted. Let's check the feature importances.
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

```python
# There is an enhanced version of GradientBoostingClassifier which is HistGradientBoostingClassifier
hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# HistGradientBoostingClassifier doesn't provide the feature importances. permutation_importance checks the importances by iterating features to the model.
hgb.fit(train_input, train_target)
result= permutation_importance(hgb, train_input, train_target, n_repeats=10, random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
# We can apply to the test set also
result= permutation_importance(hgb, test_input, test_target, n_repeats=10, random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
# Finally we can check the score for the test set
hgb.score(test_input, test_target)
```

```python
# There are also other libraries that support graident boosting. Here we test XGBoost.
xgb = XGBClassifier(tree_method='hist', random_state=42)
scores = cross_validate(xgb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# Another one is LightGBM which is made by Microsoft.
lgb = LGBMClassifier(random_state=42)
scores = cross_validate(lgb, train_input, train_target, return_train_score=True, n_jobs = -1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```
