```python
# import statements\
import keras
from sklearn.model_selection import train_test_split
```

```python
# download the data using keras
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()
```

```python
# Scale and split data
train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# Let's build nerual network model with an input and 2 dense layers.
inputs = keras.layers.Input(shape=(784,))
dense1 = keras.layers.Dense(100, activation='sigmoid')
dense2 = keras.layers.Dense(10, activation='softmax')
model = keras.Sequential([inputs, dense1, dense2])
model.summary()
```

```python
# We can also set the layer directly without declare additional variables.
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation='sigmoid', name='hidden layer'),
    keras.layers.Dense(10, activation='softmax', name='output alyer')
])
model.summary()
```

```python
# or we can also add layers one by one
model = keras.Sequential()
model.add(keras.layers.Input(shape=(784,)),)
model.add(keras.layers.Dense(100, activation='sigmoid'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()
```

```python
# Finally let's train the model.
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# Additional alyer increased the accuracy.
# We can further enhance this by applying relu activation instead of sigmoid to the desne layer.
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)),)
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()
```

```python
# We don't need to reshape since we have added the flatten layer.
train_scaled = train_input / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# Now let's train the model. We can see the relu activation increased accuracy a little bit.
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# Let's evaluate using validation set
model.evaluate(val_scaled, val_target)
```

```python
# Let's also try different optimizer. Adam uses adaptive learning rate
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)),)
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# final evaluation
model.evaluate(val_scaled, val_target)
```
